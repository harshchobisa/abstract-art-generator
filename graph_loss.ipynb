```python
import matplotlib.pyplot as plt
```

```python
f = open("output.txt", "r")
dis_losses = []
gen_losses = []
for line in f.readlines():
    if "Loss" in line:
        dis_losses.append(float(line.split("Loss_D:")[1].split("\tLoss_G")[0]))
        gen_losses.append(float(line.split("Loss_G:")[1].split("\tD(x)")[0]))
f.close()
```

```python
plt.figure(figsize=(10,5))
plt.title("Generator and Discriminator Loss During Training")
plt.plot(gen_losses,label="G")
plt.plot(dis_losses,label="D")
plt.xlabel("iterations")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
